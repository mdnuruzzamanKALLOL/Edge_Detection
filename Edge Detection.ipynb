```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
img = cv2.imread('Dataset/cat6.jpg')
```

```python
def imageMain(imgtitle,image):
    imgVer = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(5,3))
    plt.imshow((imgVer).astype(np.uint8))
    plt.title(imgtitle)
    plt.axis('off')
    plt.grid(False)
    plt.show()
```

Real Images

```python
imageMain("PussyCat", img)
```

Gray Image

```python
gray = cv2.cvtColor(img ,cv2.COLOR_RGB2GRAY)
```

```python
imageMain("Gray", gray)
```

Edge Detection ---ZERO CROSSING

```python
img = cv2.imread('Dataset/cat6.jpg',0) # 0 indicates that the image should be loaded in grayscale mode.
```

```python
blur = cv2.GaussianBlur(img, (3, 3), 0)
```

```python
laplacian = cv2.Laplacian(blur, cv2.CV_64F)
```

```python
edges = cv2.threshold(np.abs(laplacian), 30, 250, cv2.THRESH_BINARY)[1]
```

```python
plt.imshow(edges, cmap='gray')
plt.title('Zero Crossing')
plt.axis('off')
plt.show()
```

Edge Detection ---CANNY-DERICHE

```python
blurred_img = cv2.GaussianBlur(img, (5, 5), 0)
```

```python
low_threshold = 50  # Adjust this threshold based on the image and noise level
high_threshold = 150  # Typically set to a ratio of 1:3 or 1:2 to low_threshold
```

```python
edgess = cv2.Canny(blurred_img, low_threshold, high_threshold)
```

```python
imageMain("Canny–Deriche", edgess)
```
